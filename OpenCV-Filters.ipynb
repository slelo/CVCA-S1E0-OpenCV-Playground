```python
import cv2
import numpy as np
from enum import Enum
```

# Camera settings and filter category definitions
You don't need to change anything here, but feel free to change the resolution and add more filters

```python
capture = cv2.VideoCapture(0)
capture.set(3, 640)
capture.set(4, 480)

class FilterCategory(Enum):
    DEFAULT = 0
    SOBEL = 1
    CANNY = 2
    BLUR = 3
    SHARPEN = 4
    FACEDETECTION = 5

mode = FilterCategory.DEFAULT.value
```

# Playground
Play around with the filters, with the settings, try see how small changes modify the results. Refer to the OpenCV documentation when exploring the filters: https://docs.opencv.org/4.12.0/

```python
def getFilteredImg(img, mode):
    # Apply the filter based on the mode, feel free to play around
    if mode == FilterCategory.DEFAULT.value:
        pass
    elif mode == FilterCategory.SOBEL.value:
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        sobel_x = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
        sobel_y = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)
        img = cv2.addWeighted(sobel_x, 0.5, sobel_y, 0.5, 0)
    elif mode == FilterCategory.CANNY.value:
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        img = cv2.Canny(gray, 50, 200)
    elif mode == FilterCategory.BLUR.value:
        img = cv2.blur(img, (5,5))
    elif mode == FilterCategory.SHARPEN.value:
        img = cv2.filter2D(img, -1, np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]]))
    elif mode == FilterCategory.FACEDETECTION.value:
        face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(gray, 1.1, 4)
        for (x, y, w, h) in faces:
            cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)

    return img
```

This is the main loop where the filters are applied and the results are displayed

```python
while True:
    # Read the frame and check if it was successful
    success, img = capture.read()

    imgFilter = getFilteredImg(img, mode)

    cv2.imshow("Live webcam", imgFilter) # This will open an independent window

    # Wait for a key press and check if it was 'q' to quit
    key = cv2.waitKey(1) & 0xFF

    if key == ord('q') or key == 27: # Close window by pressing 'q' or ESC
        capture.release()
        break

    if ord('0') <= key <= ord('5'):
        mode = key - ord('0')

cv2.destroyAllWindows()
cv2.waitKey(1) # you might need this if you use MacOS
```
